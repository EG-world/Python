# **1. 반복문**
반복문은 동일한 작업을 여러 번 실행하기 위해 사용되는 제어 구조입니다. 주로 for 문과 while 문이 사용되며, 각각의 반복문은 다른 상황에 적합한 방식으로 사용됩니다.

# **2. while문**
while 문은 특정 조건이 참인 동안 반복적으로 코드 블록을 실행하는 제어 구조입니다. while 문은 주어진 조건이 참인 동안 반복적으로 코드를 실행하며, 조건이 거짓이 되면 반복을 멈춥니다.
```
while 조건:
    # 조건이 참인 동안 실행될 코드
```

````python
while True :  # 무한반복
  print('Hello Python')
````

````python
i = 1
while i <= 5 :
  print('Hello Python')
  i += 1
````

````python
# 1부터 10까지의 합
i = 1
sum = 0
while i<=10 :
  sum += i
  i += 1
print(sum)
````

### 문제 1
1부터 100까지의 짝수의 합을 출력하는 프로그램을 작성해보자

````python
i = 2
sum = 0
while i <= 100 :
  sum += i
  i += 2
sum
````

````python
i = 1
sum = 0
while i <= 100 :
  if i%2 == 0:
    sum += i
  i += 1
sum
````

### 문제 2
원하는 단을 입력받아 해당 단의 구구단을 출력
* 원하는 단은 입력하세요 : 원하는 단 입력
```
출력 예시)
3단
3 * 1 = 3
3 * 2 = 6
...
3 * 9 = 27
```

````python
dan = int(input('원하는 단을 입력하세요 : '))
i = 1
print(f'{dan}단')
while i < 10:
  print(f'{dan} x {i} = {dan*i}')
  i += 1
````

````python
# break : 반복문을 즉시 종료할 때 사용하는 키워드
# while 또는 for에서 모두 사용 가능
while True:
  num = int(input('짝수를 입력하세요 : '))
  if num%2==0 :
    print('짝수를 입력하셨습니다')
    print('프로그램을 종료합니다')
    break
  else :
    print('홀수를 입력하셨습니다')
    print('다시 입력해주세요')
    continue
````

# **3. for문**
for 문은 시퀀스(리스트, 튜플, 문자열 등)의 각 항목에 대해 반복 작업을 수행하는 반복문입니다. for 문은 주로 "지정된 범위 내에서 반복"할 때 사용됩니다. 이것은 반복할 요소들이 미리 정해져 있는 경우에 유용합니다.
```
for 요소 in 시퀀스:
    # 요소에 대한 작업 수행
```

````python
for i in "hello" :
  print(i)
````
